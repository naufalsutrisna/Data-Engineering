## Import Module

```python
import numpy as np
import pandas as pd
```

## Twitter

```python
df_twitter = pd.read_csv('twitter.csv', delimiter=";")
df_twitter.head()
```

```python
df_twitter['created_at'] = pd.to_datetime(df_twitter['created_at'], format='%a %b %d %H:%M:%S %z %Y')
```

```python
columns_to_drop = ['id_str', 'quote_count', 'reply_count', 'retweet_count',
                   'lang', 'user_id_str', 'conversation_id_str', 'tweet_url']

df_twitter = df_twitter.drop(columns=columns_to_drop)
```

```python
df_twitter.head()
```

## Instagram

```python
df_instagram = pd.read_csv('instagram.csv')
df_instagram.head()
```

```python
df_instagram['timestamp'] = pd.to_datetime(df_instagram['timestamp'], format='%Y-%m-%dT%H:%M:%S.%fZ')
```

```python
columns_to_drop = ['id', 'ownerProfilePicUrl', 'postUrl']

df_instagram = df_instagram.drop(columns=columns_to_drop)
```

```python
df_instagram.head()
```

## Load Dataframe

```python
columns = ['username', 'likes', 'timestamp', 'comment', 'label']
df = pd.DataFrame(columns=columns)
```

```python
df.head()
```

```python
for index in range(len(df_twitter)):
    row = df_twitter.iloc[index]
    new_row = pd.DataFrame({
        'username': [row['username']],
        'likes': [row['favorite_count']],
        'timestamp': [pd.to_datetime(row['created_at'])],
        'comment': [row['full_text']],
        'label': ['twitter']
    })
    df = pd.concat([df, new_row], ignore_index=True)
```

```python
df.shape
```

```python
for index in range(len(df_instagram)):
    row = df_instagram.iloc[index]
    new_row = pd.DataFrame({
        'username': [row['ownerUsername']],
        'likes': [row['likesCount']],
        'timestamp': [pd.to_datetime(row['timestamp'])],
        'comment': [row['text']],
        'label': ['instagram']
    })
    df = pd.concat([df, new_row], ignore_index=True)
```
